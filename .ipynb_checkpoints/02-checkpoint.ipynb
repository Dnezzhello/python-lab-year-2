```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
```

```python
df = pd.read_csv("DigiDB_digimonlist.csv")
```

```python
df.head()
```

```python
# count nulls in each feature
df.isnull().sum()
```

```python
# deal with those nulls
for col in df.columns:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
# statistical summary
df.describe()
```

```python
# find out how many digimons in each stage
df.Stage.value_counts()
```

```python
# group data by "Type" and find mean for each remaining feature
df.groupby(by="Type").mean()
```

```python
# sort the data
df.sort_values(by="Lv 50 HP", ascending=False).head()
```

```python
# select rows based on given conditions
df[df["Type"] == "Virus"]
```

```python
# crate a new column that sums Lv50 Sp, Lv50 Atk, Lv50 Def, Lv50 Int and Lv50 Spd up

df["Total_status"] = df[["Lv50 SP", "Lv50 Atk", "Lv50 Def", "Lv50 Int", "Lv50 Spd"]].sum(axis=1)
```

```python
df.head()
```

```python
# intentionally skip merging table part
```

```python
# histogram
df["Total_status"].hist();
```

```python
status_cols = ["Lv50 SP", "Lv50 Atk", "Lv50 Def", "Lv50 Int", "Lv50 Spd"]

plt.figure(figsize=(10,8), dpi=150)
for col in status_cols:
    df[col].hist(alpha=0.5)
    
plt.legend(status_cols)
plt.show()
```

```python
# scatter plot

plt.figure(figsize=(10,8), dpi=150)
plt.scatter(data=df, x="Lv50 Atk", y="Lv50 Def", c="Memory")
plt.xlabel("Lv50 Atk")
plt.ylabel("Lv50 Def")
plt.colorbar()
plt.show()
```

```python
# box plot
plt.figure(figsize=(10,8), dpi=150)
df[status_cols].boxplot()
plt.show()
```
